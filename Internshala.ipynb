```python
# Settings
output_file = 'synthetic_data.csv'
target_size_gb = 1  # Target size of the dataset
sampling_rate_hz = 10
duration_seconds = 3600 * 24  # Simulating 1 day (adjust if necessary)
```

```python
# Calculate samples based on size target
estimated_row_size = 200  # Approximate bytes per row
rows_needed = int((target_size_gb * 1e9) / estimated_row_size)
```

```python
# Initialize data storage
timestamps = []
cpu_temperatures = []
cpu_usages = []
cpu_loads = []
memory_usages = []
battery_levels = []
cpu_powers = []
```

```python
# Generate synthetic data
for i in range(rows_needed):
    current_time = datetime.datetime.now()
    timestamps.append(current_time)

    # Randomly generate synthetic sensor data
    cpu_temperatures.append(random.uniform(20, 100))  # Normal + anomaly range
    cpu_usages.append(random.uniform(0, 100))
    cpu_loads.append(random.uniform(0, 5))
    memory_usages.append(random.uniform(0, 100))
    battery_levels.append(random.uniform(0, 100))
    cpu_powers.append(random.uniform(0, 150))  # Normal + anomaly range

    # Periodically save to CSV to avoid memory issues
    if i % 10000 == 0 or i == rows_needed - 1:
        data_chunk = pd.DataFrame({
            'timestamp': timestamps,
            'cpu_temperature': cpu_temperatures,
            'cpu_usage': cpu_usages,
            'cpu_load': cpu_loads,
            'memory_usage': memory_usages,
            'battery_level': battery_levels,
            'cpu_power': cpu_powers
        })
        data_chunk.to_csv(output_file, mode='a', index=False, header=not os.path.exists(output_file))
        timestamps.clear()
        cpu_temperatures.clear()
        cpu_usages.clear()
        cpu_loads.clear()
        memory_usages.clear()
        battery_levels.clear()
        cpu_powers.clear()

print("Data generation complete!")
```

```python
import numpy as np
from scipy.stats import zscore
# Calculate the z-score for CPU temperature
df['cpu_temperature_zscore'] = zscore(df['cpu_temperature'])

lower_bound = df['cpu_temperature'].quantile(0.05)
upper_bound = df['cpu_temperature'].quantile(0.95)
df['is_anomaly'] = (df['cpu_temperature'] < lower_bound) | (df['cpu_temperature'] > upper_bound)
print(f"Total anomalies detected using percentiles: {df['is_anomaly'].sum()}")

```
